```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('datasets_student-por.csv')
data
```

```python
data.columns == data.columns
data.columns
```

```python
obj = list()
num = list()

for n in data.columns:
    if data[n].dtype == 'int64':
        num.append(n)
    else:
        obj.append(n)
```

```python
print(obj) 
print(num)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(),annot=True)
plt.yticks(rotation=0)
```

```python
correlated_data =[]
corr = data[num].corr()
for i in num:
    for j in num:
        if corr[i][j]>0.80 and i!=j:
            correlated_data.append(i)
            correlated_data.append(j)
correlated_data = list(set(correlated_data))
    
```

```python
print(correlated_data)
```
